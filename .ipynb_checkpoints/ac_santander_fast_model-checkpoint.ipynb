```python
# 1. Import libraries


import os
import time
#import warnings

# Base libraries
import numpy as np
import pandas as pd

# Wrappers
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

# Preprocessing
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, MinMaxScaler, FunctionTransformer

# Training and validation
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV

# Models
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from xgboost.sklearn import XGBClassifier

# Metrics
# Classification
# from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
# from sklearn.metrics import recall_score
# from sklearn.metrics import precision_score
# from sklearn.metrics import precision_recall_curve
# from sklearn.metrics import roc_curve

# Regression
# from sklearn.metrics import r2_score
# from sklearn.metrics import mean_squared_error
# from sklearn.metrics import mean_absolute_error


# dump and load models
# import joblib

# df display options
pd.set_option('display.max_columns', False) # Show all the features
# pd.set_option('display.max_rows', False) # Show all the instances
```

```python
# 2. Import datasets
raw_data = pd.read_csv('/home/shared/santander/EVS_KAGGLE_TEAM/datasets/train.csv')
raw_test = pd.read_csv('/home/shared/santander/EVS_KAGGLE_TEAM/datasets/test.csv')
```

```python
raw_data.head()
```

```python
# Separate response variable

X = raw_data.drop(['TARGET','ID'], axis=1)
y = raw_data.TARGET
```

```python
# Create group of features

num_cols = X.columns.values
```

```python
# 3. Data preprocessing
num_proc = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value = -1))
])

preprocessor = ColumnTransformer(transformers=[
    ('num', num_proc, num_cols)
])
```

```python
# 4. Modeling
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state = 1234)
```

```python
# Baseline

baseline = Pipeline(steps=[
    ('preprocessing', preprocessor)
    ,('classifier', DummyClassifier(random_state = 1234))
])

baseline.fit(X_train, y_train)
```

```python
# Performance metrics
print('Train set score: {:.4f}'.format(roc_auc_score(y_train, baseline.named_steps['classifier'].predict(X_train))))
print('Train set score: {:.4f}'.format(roc_auc_score(y_test, baseline.named_steps['classifier'].predict(X_test))))
```

```python
# Find a better model

models = [
            {'classifier':[LogisticRegression(random_state = 1234)], 'preprocessing':[preprocessor]}
    
            ,{'classifier':[XGBClassifier(random_state = 1234)], 'preprocessing':[preprocessor]
              
              }]
```

```python
# Cross-validation
k=5 # Number of folders to use
grid = GridSearchCV(baseline, param_grid = models, cv=k, scoring='roc_auc')
grid.fit(X_train, y_train)
```

```python
# Best Model
champ = grid.best_estimator_.named_steps['classifier']
print('Best model {}'.format(champ))
```

```python
# New metrics
print('Best cross-validation score: {:.4f}'.format(grid.best_score_))
print('Train set score: {:.4f}'.format(grid.score(X_train, y_train)))
print('Test set score: {:.4f}'.format(grid.score(X_test, y_test)))
```
