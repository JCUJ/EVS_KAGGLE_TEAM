```python
#1. Import essential libraries

# Timer
import os
import time

# Base libraries
import numpy as np
from numpy import loadtxt
import pandas as pd

# Keras
from keras.models import Sequential
from keras.layers import Dense

import matplotlib.pyplot as plt
```

```python
# 2. Import dataset
train = np.genfromtxt ('C:/Users/Alejandro/Documents/Kaggle/train.csv', delimiter=",")
test = np.genfromtxt ('C:/Users/Alejandro/Documents/Kaggle/train.csv', delimiter=",")
```

```python
# 3. Split
X = train[1:,1:-1]
y = train[1:,-1]
```

```python
X.shape
```

```python
# 4. Define NN
model = Sequential()
model.add(Dense(12, input_dim=369, activation='relu'))
model.add(Dense(369, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# 5. Fit
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X, y, validation_split=0.2, epochs=10, batch_size=120)
```

```python
# 6. Performance
accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
print('Training set accuracy:')
print(np.round(accuracy,4))
print('Validation set accuracy:')
print(np.round(val_accuracy,4))
```
